```python
# Numpy Practice Problem
# Crop and Rotate Image using Numpy
import cv2
import numpy as np
import matplotlib.pyplot as plt


im_path = "img2.png"
im_array = cv2.imread(im_path)
print("original")
plt.imshow(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB))
plt.show()


# this image has 512 pixels width and 512 pixels height
# print(im_array.shape)
# crop the image using numpy 10 percent from top, 20 percent from bottom
# 5 percent from left and 25 percent from right
# left = np.percentile(im_array, 5)
# right = np.percentile(im_array, 25)
# top = np.percentile(im_array, 10)
# bottom = np.percentile(im_array, 20)
height = im_array.shape[0]
width = im_array.shape[1]
top, bottom, left, right = .1, .2, .05, .25
# cropped_im_array = your code here

print("x - start :",0+int(height*top))
print("x - end :",height-int(height*bottom))
print("y - start :",0+int(width*left))
print("y - end :",width-int(width*right))

cropped_im_array = im_array[0+int(height*top):height-int(height*bottom),
                            0+int(width*left):width-int(width*right)]
  
# Convert array to image
# im_path = im_path.fromarray(croped)


print("cropped")
plt.imshow(cv2.cvtColor(cropped_im_array, cv2.COLOR_BGR2RGB))
plt.show()


# rotate this image on 90 degree
# rotated_im_array = your code here
degrees = 90
rotated_im_array = im_array.copy()
for _ in range(degrees // 90):
    rotated_im_array = rotated_im_array.transpose(1, 0, 2)
    for j in range(0, rotated_im_array.shape[1] // 2):
        c = rotated_im_array[:, j, :].copy()
        rotated_im_array[:, j, :] = rotated_im_array[: , rotated_im_array.shape[1]-j-1, :]
        rotated_im_array[: , rotated_im_array.shape[1]-j-1, :] = c

print("rotated")
plt.imshow(cv2.cvtColor(rotated_im_array, cv2.COLOR_BGR2RGB))
plt.show()
# Expected input and output is attached
```
